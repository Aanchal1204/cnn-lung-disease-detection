```python
import numpy as np
import matplotlib.pyplot as plt

# Classes
classes = ['Bacterial Pneumonia', 'Corona Virus', 'Normal', 'Tuberculosis', 'Viral Pneumonia']
x = np.arange(len(classes))
width = 0.25

# ---- Overall Accuracy (single value per model) ----
accuracy = {
    "EfficientNet": 0.856,
    "DenseNet": 0.845,
    "ResNet": 0.53
}

# ---- Class-wise Metrics (replace with your real values) ----
precision = {
    "EfficientNet": [0.729, 0.95, 0.87, 0.98, 0.69],
    "DenseNet":     [0.73, 0.95, 0.86, 0.97, 0.69],
    "ResNet":       [0.52,0.61,0.45,0.68,0.48],
}

recall = {
    "EfficientNet": [0.65, 0.96, 0.96, 0.95, 0.76],
    "DenseNet":     [0.67, 0.95, 0.97, 0.98, 0.67],
    "ResNet":       [0.36,0.71,0.93,0.5,0.15],
}

f1 = {
    "EfficientNet": [0.71, 0.95, 0.91, 0.97, 0.73],
    "DenseNet":     [0.70, 0.96, 0.91, 0.97, 0.68],
    "ResNet":       [0.43, 0.65, 0.61, 0.58, 0.23],
}

# ---- Create 2x2 layout ----
fig, axs = plt.subplots(2, 2, figsize=(16, 10))
axs = axs.flatten()

# ---------------- Accuracy Plot ----------------
ax = axs[0]
models = list(accuracy.keys())
values = list(accuracy.values())

ax.bar(models, values)
ax.set_title("Overall Accuracy")
ax.set_ylim(0, 1)
ax.set_ylabel("Score")

# ---------------- Precision Plot ----------------
ax = axs[1]
ax.bar(x - width, precision["EfficientNet"], width, label='EfficientNet')
ax.bar(x, precision["DenseNet"], width, label='DenseNet')
ax.bar(x + width, precision["ResNet"], width, label='ResNet')

ax.set_title("Precision (Class-wise)")
ax.set_xticks(x)
ax.set_xticklabels(classes, rotation=20)
ax.set_ylim(0, 1)
ax.legend()

# ---------------- Recall Plot ----------------
ax = axs[2]
ax.bar(x - width, recall["EfficientNet"], width, label='EfficientNet')
ax.bar(x, recall["DenseNet"], width, label='DenseNet')
ax.bar(x + width, recall["ResNet"], width, label='ResNet')

ax.set_title("Recall (Class-wise)")
ax.set_xticks(x)
ax.set_xticklabels(classes, rotation=20)
ax.set_ylim(0, 1)
ax.legend()

# ---------------- F1 Score Plot ----------------
ax = axs[3]
ax.bar(x - width, f1["EfficientNet"], width, label='EfficientNet')
ax.bar(x, f1["DenseNet"], width, label='DenseNet')
ax.bar(x + width, f1["ResNet"], width, label='ResNet')

ax.set_title("F1 Score (Class-wise)")
ax.set_xticks(x)
ax.set_xticklabels(classes, rotation=20)
ax.set_ylim(0, 1)
ax.legend()

# ---- Final Layout ----
plt.tight_layout()
plt.show()
```
